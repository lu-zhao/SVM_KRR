```python
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
filename = os.path.join('data', "Sodermalmsallen4-6.csv")
df = pd.read_csv(filename,sep=',')
df = df.set_index('Date_Time')
df.head()
```

# Correlations among features

```python
df1 = df[['Consumption', 'hour','weekday','weekday_','one_hour_before','two_hour_before',
         'three_hour_before','1days_before','2days_before','7days_before','weather']]
```

```python
plt.matshow(df1.corr(method='spearman'),vmax=1,vmin=-1,cmap='PRGn')
# plt.title('without resampling', size=15)
plt.colorbar()
plt.show()
```
